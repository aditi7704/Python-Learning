### 2.4 Boolean:

#### 2.4.1. bool

#### A Boolean represents one of two possible values:
##### True
##### False
##### These are used to express truth values and are extremely important in conditions, comparisons, loops, and logical operations.
##### In Python : They start with a capital T and capital F.

```python
True
False
```

##### Boolean is a Subclass of int
###### In Python, True and False are actually integers:
###### True acts like 1
###### False acts like 0
###### Example:

```python
print(True + True)   # 2
print(True + False)  # 1
print(False + False) # 0
print(isinstance(True, int))  # True
```

##### Where Do Booleans Come From?
###### Booleans often come from comparisons or logical expressions.

```python
# comparisions
print(10 > 5)   # True
print(10 < 5)   # False
print(5 == 5)   # True
print(5 != 2)   # True
```

```python
print(True and False) # False
print(True or False)  # True
print(not True)       # False
```

#### Type Conversion to Bool (bool())
##### Python automatically converts many values to booleans using bool():
###### These become False :
###### 0
###### 0.0
###### "" (empty string)
###### [] (empty list)
###### {} (empty dict)
###### () (empty tuple)
###### None

```python
print(bool(0))     # False
print(bool(""))    # False
print(bool(None))  # False
print(bool(5))      # True
print(bool("Hi"))   # True
print(bool([1,2]))  # True
```

#### Booleans in if-else 

```python
#Booleans are most often used in decision making:
x = 10
if x > 5:
    print("x is greater")
else:
    print("x is smaller")
```

### 2.5 Set type:
#### Both are used to store unique, unordered collections of elements.

#### 2.5.1. Set

#### 2.5.1. FrozenSet
##### Same as set (unique, unordered)
##### Immutable → cannot add/remove items
##### Useful when:
###### You want a constant set
###### You want to store a set inside another set or dictionary key

```python
fs = frozenset([10, 20, 30])
print(fs)                  # frozenset({10, 20, 30})
# fs.add(40)  ❌ Error (because immutable)
# fs.remove(10) ❌ Error
```

```python
#You can still use mathematical operations:
A = frozenset([1, 2, 3])
B = frozenset([3, 4, 5])
print(A.union(B))          # frozenset({1, 2, 3, 4, 5})
print(A.intersection(B))   # frozenset({3})
```
