# 2.Find the maximum and minimum element in a list without using max() or min().

```python
l=[1,2,3,4,5]
max=max(l)
min=min(l)
print(max)
print(min)
```

```python
l=[8,2,1,2,4,6,5,0,-1]
max=l[0]
for i in l:
    if max <= i:
        max=i
print(max)
min=l[0]
for i in l:
    if min >= i:
        min=i
print(min)
```

# 3.Count how many times a specific element appears in a list.

```python
l=[1,2,3,4,5,5,6,3,2]
n=int(input("Enter Element from the list"))
print(l.count(n))
```

```python
l=[1,2,3,4,5,5,6,3,2]
n=int(input("Enter Element from the list"))
count=0
for i in l:
    if n==i:
        count+=1
print(count)
```
