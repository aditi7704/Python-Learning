# Type Casting

## 1. Definition:
### 1.1. Type Casting in Python means converting one data type into another.
### 1.2. Python is dynamically typed, so the interpreter automatically decides data types, but sometimes we manually convert types—this is called explicit type casting.

## 2. Types:
### 2.1 Implicit Type Casting
### 2.2 Explicit Type Casting

### 2.1 Implicit Type Casting
#### Performed automatically by Python
#### Happens when a smaller/compatible data type is converted into a larger one
#### No data loss

```python
x = 10       # int
y = 2.5      # float
z = x + y    # Python automatically converts x to float
print(z)     # 12.5
print(type(z))
#Why float?
#Because:
#int → float is safe
#float can store more complex values
#Python always promotes to the “higher” type
```

### 2.2 Explicit Type Casting
#### We convert data types manually using built-in functions.
#### Common Type Casting Functions
#### Function	Converts To
#### int()	Integer
#### float()	Floating number
#### complex()	Complex number
#### str()	String
#### bool()	Boolean
#### list()	List
#### tuple()	Tuple
#### set()	Set

#### 2.2.1 Casting to Integer → int()

##### 2.2.1.1. float to int

```python
x = 12.9
print(int(x))   # 12 (decimal part removed)
```

##### 2.2.1.2. String to int

```python
s = "100"
print(int(s))   # 100
```

#### 2.2.2 Casting to Float → float()

```python
x = "5"
print(float(x))   # 5.0
```

#### 2.2.3. Casting to String → str()

```python
x = 100
print(str(x))   # "100"
```

#### 2.2.4. Casting to Boolean → bool()
##### Rules:
##### 0, 0.0, empty values ("", [], {}) → False
##### Everything else → True

```python
print(bool(0))      # False
print(bool(15))     # True
print(bool(""))     # False
print(bool("Hi"))   # True
```

#### 2.2.5. Casting to List / Tuple / Set

```python
s = "ABC"
print(list(s))   # ['A', 'B', 'C']
print(tuple(s))  # ('A', 'B', 'C')
print(set(s))    # {'A', 'B', 'C'}
```

# Interview-Ready Answer (Short Version)
## Type casting in Python is the process of converting one data type into another.
## Python supports two types:
### Implicit Type Casting – done automatically by Python, usually converting lower precision types (int) into higher (float) to avoid data loss.
### Explicit Type Casting – done manually using functions such as int(), float(), str(), bool(), list(), etc.
### Type casting is important for operations where data types must match or when taking input from the user (which is always a string).
